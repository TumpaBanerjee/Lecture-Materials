# Visualize Continuos data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data1=pd.read_table('C:\MY DRIVE\Personal\dataset\All_Data\All_Data\P088.txt',sep='\t')
```

```python
data1
```

```python
data1.info()
```

```python
plt.plot(data1['Age'])
plt.title('Age')
```

```python
fig=plt.figure()
ax1=fig.add_subplot(2,2,1)
ax1.plot(data1['Age'])
ax2=fig.add_subplot(2,2,2)
ax2.plot(data1['HS'])
ax3=fig.add_subplot(2,2,3)
ax3.plot(data1['Income'])
ax4=fig.add_subplot(2,2,4)
ax4.plot(data1['Black'])
```

# Check Data distribution

```python
fig=plt.figure()
ax1=fig.add_subplot(2,2,1)
ax1.hist(data1['Age'],bins=10,color='red')
ax2=fig.add_subplot(2,2,2)
ax2.hist(data1['HS'],bins=10,color='pink')
ax3=fig.add_subplot(2,2,3)
ax3.hist(data1['Income'],bins=10,color='green')
ax4=fig.add_subplot(2,2,4)
ax4.hist(data1['Black'],bins=10,color='blue')
```

# Plot all the data in one figure

```python
fig=plt.figure()
ax1=fig.add_subplot(1,1,1)
ax1.plot(data1['Age'],'k',label='Age',color='red')
ax1.plot(data1['HS'],'k--',label='HS',color='blue')
ax1.plot(data1['Black'],'k--',label='HS',color='green')
ax1.legend(loc='best')
```

# Scatter Plot

```python
plt.scatter(data1['Age'],data1['Sales'])
plt.title('Scatter Plot')
plt.xlabel('Age')
plt.ylabel('Sales')
```
